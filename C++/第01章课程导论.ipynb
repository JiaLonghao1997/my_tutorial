## 1-3 计算机语言和程序设计方法的发展<br>
机器语言：<br>
	• 由二进制代码构成<br>
	• 计算机硬件可以识别<br>
	• 可以进行简单的操作<br>
	• 例如：加法、减法、数据移动等等<br>
最初的计算机语言-机器语言， 和人类的自然语言之间存在巨大鸿沟。<br>
汇编语言：<br>
	• 将机器指令映射为一些助记符。如：ADD、SUB、mov等。<br>
	• 抽象层次低，需要考虑机器细节。<br>
高级语言<br>
	• 关键字、语句容易理解；<br>
	• 有含义的数据命名和算式；<br>
	• 抽象层次较高，例如，算式：a+b+c/d<br>
	• 屏蔽了机器底层的细节；例如，这样显示计算结果：count<<a+b+c/d<br>
C++语言<br>
	• 是高级语言<br>
	• 支持面向对象的观点和方法<br>
		○ 将客观事物看做对象；<br>
		○ 对象间通过消息传送进行沟通；<br>
		○ 支持分类和抽象；<br>
程序设计的方法和发展历程<br>
面向过程的程序设计方法：<br>
	• 机器语言、汇编语言、高级语言都支持<br>
	• 最初目的：由于数学计算<br>
	• 主要工作：设计求解问题的过程<br>
大型复杂的软件，难以用面向过程的方式编写。<br>
面向对象的程序设计方法：<br>
	• 由面向对象的高级语言支持<br>
	• 一个系统由对象构成<br>
	• 对象与对象之间通过消息进行通信。<br>
## 1-4面向对象的基本概念<br>
对象（object)：<br>
	• 一般意义上的对象：是现实世界中一个世纪存在的事物；<br>
	• 面向对象方法中的对象：是系统用来描述客观事物的一个实体。<br>
抽象与分类<br>
	• 分类所依据的原则——抽象<br>
	• 抽象出同一类对象的共同属性和行为，形成类；<br>
	• 类与对象的关系：类型和实例的关系；犹如模具与铸件之间的关系。<br>
封装：<br>
	• 隐蔽对象的内部细节<br>
	• 对外形成一个边界<br>
	• 只保留有限的对外接口<br>
	• 使用方便、安全性好<br>
继承<br>
	• 意义在于软件复用；<br>
	• 改造、扩展已有类形成新的类<br>
多态(Polymorphism)<br>
	• 同样的消息作用在不同对象上，有可能引起不同的行为；<br>
    
## 1-5 程序的开发过程<br>
高级语言——>机器语言程序：需要进行翻译<br>
源程序：有源语言写的，有待翻译的程序；<br>
目标程序：源程序通过翻译加工以后生成的机器语言程序；<br>
可执行程序：源程序+某块（连接）后形成可执行程序<br>
	• 连接目标程序以及库中的某些文件，生成的一个可执行文件。例如：Windows系统平台上的.EXE文件。<br>
三种不同类型的翻译程序：<br>
	• 汇编程序：将汇编语言源程序翻译成目标程序；<br>
	• 编译程序：将高级语言源程序翻译成目标程序；<br>
	• 解释程序：将高级语言源程序翻译成机器指令，它是边翻译边执行。<br>
	                  ——解释一条程序，执行一条程序（每次都需要）。<br>
问题：解释程序执行效率比较低，为什么需要这种方式呢？<br>
Java程序就是半编译半解释的，目的是为了跨平台。——二进制代码+Java虚拟机<br>
C++程序是直接编译成本地机器语言代码。<br>
C++程序的开发过程<br>
	• 算法与数据结构设计；<br>
	• 源程序编辑<br>
	• 编译<br>
	• 连接<br>
	• 测试（各种数据，测试）<br>
	• 调试（找到错误原因+修改错误）<br>

## 1-6-3数据的编码表示<br>
二进制数都可以机器识别了，为什么还需要编码呢?<br>
负数如何表示<br>
最容易想到的方案：0表示“+”号， 1表示“-”号。<br>
原码：“符号--绝对值”表示的编码<br>
原码的缺点：<br>
	• 零的表示不唯一；<br>
	• 进行四则运算时，符号位必须单独处理且运算规则复杂。<br>
补码：<br>
	• 0的表示唯一<br>
	• 符号位可以作为数值参加运算<br>
	• 减法运算可以转化为加法运算<br>
模数：n位二进制整数的模数为“2”，n位小数的模数为2；<br>
补数：一个数减去另一个数（加一个负数）等于第一个数加上第二个数的补数。<br>
(计算机组成原理——补码的原理）<br>
例如：始终指针 8+（-2）=8+10（mod 12) = 6<br>

补码的计算规则<br>
反码：作为中间码（本身没什么用）<br>
反码的计算规则：<br>
1.负整数：原码的符号位不变（仍是1），其余各位取反（0变1，1变0）。<br>
2.正整数：原码就是补码；<br>
补码的计算规则：<br>
反码：作为中间码<br>
负数补码 = 反码 +1 <br>
正数补码 = 原码<br>
补码的有点：<br>
	• 0的表示唯一；<br>
	• 符号位可以作为数值参加运算；<br>
	• 补码运算的结果仍为补码；<br>
如果负数之和得正数或正数之和得负数，说明计算结果溢出。<br>
（大数运算）<br>
小数的表示：<br>
	定点数：总共32位，设定了某个位置为小数点；<br>
	浮点数：<br>
	实数的浮点表示：计算机中通常采用浮点方式表示小数；<br>
字符在计算机中的表示：<br>
	• 字符在计算机中通过编码表示；<br>
	ASCII码； 汉字编码；<br>
